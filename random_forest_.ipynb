```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/car_evaluation.csv')
```

```python
df.head()
```

```python
col_names = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']


df.columns = col_names

col_names
```

```python
df.buying.value_counts()
```

```python
df.buying = df.buying.map({'vhigh':3,'high':2,'med':1,'low':0})
df.head()
```

```python
df.buying = df.buying.astype(int)
df.info()
```

```python
df.maint.value_counts()
```

```python
df.maint = df.maint.map({'vhigh':3,'high':2,'med':1,'low':0})
df.head()
```

```python
df.info()
```

```python
df.doors.value_counts()
```

```python
df.doors = df.doors.map({'3':0,'4':1,'5more':2,'2':3})
df.head()
```

```python
df.doors = df.doors.astype(int)
df.info()
```

```python
df.persons.value_counts()
```

```python
df.persons = df.persons.map({'4':2,'more':1,'2':0})
df.persons = df.persons.astype(int)
df.info()
df.head()

```

```python
df.lug_boot.value_counts()
```

```python
df.lug_boot = df.lug_boot.map({'big':2,'med':1,'small':0})
df.lug_boot = df.lug_boot.astype(int)
df.info()
df.head()

```

```python
df.safety.value_counts()
```

```python
df.safety = df.safety.map({'high':2,'med':1,'low':0})
df.safety = df.safety.astype(int)
df.info()
df.head()

```

```python
df['class'].value_counts()
```

```python
df['class'] = df['class'].map({'unacc':3,'acc':2,'vgood':1,'good':0})
df["class"] = df["class"].astype(int)
df.info()
df.head()
```

```python
X = df.drop(['class'], axis=1)

y = df['class']
```

```python
# split data into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
X_train.shape
X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier



# instantiate the classifier

rfc = RandomForestClassifier(random_state=0)



# fit the model

rfc.fit(X_train, y_train)


```

```python
from sklearn.metrics import accuracy_score
y_predict = rfc.predict(X_test)
accuracy_score(y_test,y_predict)
```

```python
#for default 10 trees we got accuracy 95#
```

```python
from sklearn.ensemble import RandomForestClassifier



# instantiate the classifier

rfc2 = RandomForestClassifier(n_estimators=100, random_state=0)



# fit the model

rfc2.fit(X_train, y_train)


```

```python
from sklearn.metrics import accuracy_score
y_predict1 = rfc2.predict(X_test)
accuracy_score(y_test,y_predict1)
```

```python
# declare feature vector and target variable

X = df.drop(['class', 'doors'], axis=1)

y = df['class']
```

```python
# split data into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier



# instantiate the classifier

rfc3 = RandomForestClassifier(random_state=0)



# fit the model

rfc3.fit(X_train, y_train)


```

```python
from sklearn.metrics import accuracy_score
y_predict2 = rfc3.predict(X_test)
accuracy_score(y_test,y_predict2)
```
